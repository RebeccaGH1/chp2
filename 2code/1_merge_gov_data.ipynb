```python
import pandas as pd

# -------------------------
# 1. Abrir arquivos
# -------------------------
file1 = r"G:\My Drive\PhD\proposal\code\chp2\data\merge.xlsx"
file2 = r"G:\My Drive\PhD\proposal\code\chp2\data\Intervencoes PAC UAP_19fev26.xlsx"

df1 = pd.read_excel(file1)
df2 = pd.read_excel(file2)
```

```python
# -------------------------
# 2. Padronizar colunas chave
# -------------------------

# Garantir que são string
df1['contrato'] = df1['Contrato'].astype(str)
df2['contrato'] = df2['Contrato'].astype(str)

df1['municipio'] = df1['Município'].astype(str)
df2['municipio'] = df2['Município'].astype(str)

df1['UF'] = df1['UF'].astype(str)
df2['UF'] = df2['UF'].astype(str)

# Remover zeros à esquerda no contrato
df1['contrato'] = df1['contrato'].str.lstrip('0')
df2['contrato'] = df2['contrato'].str.lstrip('0')

# Padronizar municipio (opcional mas recomendado)
df1['municipio'] = df1['municipio'].str.strip().str.upper()
df2['municipio'] = df2['municipio'].str.strip().str.upper()

df1['UF'] = df1['UF'].str.strip().str.upper()
df2['UF'] = df2['UF'].str.strip().str.upper()

# -------------------------
# 3. Merge
# -------------------------

df_merged = df1.merge(
    df2,
    on=['contrato', 'municipio', 'UF'],
    how='left',
    suffixes=('', '_df2')
)
```

```python
to_drop = [
    'Entregou Relatório de Avaliação?',
    'Município_df2',
    'Contrato_df2',
    'Contrato'  # você pediu drop Contrato
]
df_merged = df_merged.drop(columns=[c for c in to_drop if c in df_merged.columns], errors='ignore')
```

```python
# ---------------------------------------------------
# 2) Função para "resolver duplicadas"
# ---------------------------------------------------
def is_missing(x):
    # considera NaN, string vazia e 'nan'
    if pd.isna(x):
        return True
    if isinstance(x, str) and x.strip() == "":
        return True
    if isinstance(x, str) and x.strip().lower() == "nan":
        return True
    return False

def reconcile_pair(df, col_a, col_b, drop_if_equal=True):
    """
    Regras:
    - se A vazio e B tem valor -> A = B
    - se B vazio e A tem valor -> B = A (opcional, mas você pediu "copiar o valor da outra nela")
    - se ambos têm valor:
         - se iguais -> drop col_b (se drop_if_equal)
         - se diferentes -> mantém ambos
    Retorna df e um flag dizendo se dropou col_b
    """
    if col_a not in df.columns or col_b not in df.columns:
        return df, False

    # preencher vazios em ambos os sentidos
    a = df[col_a]
    b = df[col_b]

    # A <- B quando A vazio
    mask_a = a.map(is_missing) & (~b.map(is_missing))
    df.loc[mask_a, col_a] = df.loc[mask_a, col_b]

    # B <- A quando B vazio
    mask_b = b.map(is_missing) & (~df[col_a].map(is_missing))
    df.loc[mask_b, col_b] = df.loc[mask_b, col_a]

    # checar igualdade quando ambos têm valor (comparação como string “limpa”)
    a2 = df[col_a].astype(str).str.strip()
    b2 = df[col_b].astype(str).str.strip()

    both_present = (~df[col_a].map(is_missing)) & (~df[col_b].map(is_missing))
    equal_mask = both_present & (a2 == b2)

    # se todas as linhas com ambos presentes são iguais => pode dropar
    drop_b = False
    if drop_if_equal:
        any_both = both_present.any()
        all_equal_where_both = (equal_mask[both_present]).all() if any_both else True
        # se não existe caso "diferente" quando ambos presentes, drop
        if all_equal_where_both:
            df = df.drop(columns=[col_b])
            drop_b = True

    return df, drop_b
```

```python
# ---------------------------------------------------
# 3) Reconciliar pares repetidos
# ---------------------------------------------------
pairs = [
    ('Fonte', 'Fonte_df2'),
    ('Nome da Intervenção', 'Nome da Intervenção_df2'),
    ('Modalidade', 'Modalidade_df2'),
    # você também citou Contrapartida vs Valor de Contrapartida, Investimento vs Valor de Investimento
    # mas aqui são conceitos diferentes (2017 vs 2025). Então eu NÃO dropo automaticamente.
]

for a, b in pairs:
    df_merged, _ = reconcile_pair(df_merged, a, b, drop_if_equal=True)

# ---------------------------------------------------
# 4) Renomear colunas (como você pediu)
# ---------------------------------------------------
rename_map = {
    'Repasse / Valor de Empréstimo': 'repasse_2017',
    'Ano Seleção': 'selec_year_2017',
    'Liberado VE/VR': 'liberado_2017',
    'Valor Liberado até dez/25': 'liberato_dez2025',
    'Investimento': 'valor_inv_2017',
    'Valor de Investimento ': 'valor_inv_2025',   # atenção: pode ter espaço no fim
    'Andamento': 'andamento_2017',
    '% Realizado': 'perc_done_2017',
    'Data/ Previsão de término da obra ': 'prev_end_2025',
    'Status Contrato_df2': 'status_contr_2025',
    'Status Contrato': 'status_contr_2027',
    'Situação Obra': 'sit_obra_2017',
    'Início Efetivo ': 'effec_beg_2025',
    'Data de Assinatura': 'data_ass_2017',
    'Situação Obra ': 'sit_obra_2025',
    'Valor de Repasse  ': 'valor_repasse_2025',
    'Valor de Contrapartida  ': 'valor_contrap_2025',
    'Valor de Investimento ': 'valor_inv_total_2025',
}
# aplicar só nas colunas existentes
rename_map_existing = {k: v for k, v in rename_map.items() if k in df_merged.columns}
df_merged = df_merged.rename(columns=rename_map_existing)
```

```python

# ---------------------------------------------------
# 5) Reordenar colunas para ficar "lado a lado"
#    (2017 perto de 2025/2027, como você pediu)
# ---------------------------------------------------
preferred_order = [
    # chaves / ids
    'IBGE-7', 'UF', 'Município', 'municipio', 'contrato', 'source',

    # identificação da obra
    'Pac 1 ou Pac 2', 'Proponente', 'Programa', 'Tipo', 'Eixo Habitação',
    'Origem Recurso', 'Classificação Funcional-Programática',
    'Nome da Intervenção', 'Local Intervenção', 'Modalidade', 'Tomador',

    # status lado a lado
    'andamento_2017', 'sit_obra_2017', 'sit_obra_2025',
    'status_contr_2025', 'status_contr_2027',
    'perc_done_2017',

    # datas próximas
    'data_ass_2017', 'effec_beg_2025', 'prev_end_2025',

    # valores (lado a lado, como você pediu)
    'liberado_2017', 'repasse_2017', 'Contrapartida', 'valor_inv_2017',
    'valor_repasse_2025', 'valor_contrap_2025', 'valor_inv_total_2025', 'liberato_dez2025',

    # demais
    'Famílias Beneficiadas', 'Ano Seleção'
]

# manter só as que existem, e depois acrescentar as restantes no final
existing_pref = [c for c in preferred_order if c in df_merged.columns]
remaining = [c for c in df_merged.columns if c not in existing_pref]
df_merged = df_merged[existing_pref + remaining]
```
